**Section 3: Indexing & Slicing**

```
import numpy as np
#21.Given data = np.arange(8), what is data[2:7]?
data = np.arange(8)
print(data)
print("\n")
print("Data [2:7] is:\n",data[2:7])
#The data[2:7] is [2 3 4 5 6]
print("\n")

#22.Given m = np.arange(8).reshape(2,4), what is m[1, 2]?
m = np.arange(8)#0,1,2,3,4,5,6,7
re_sp=m.reshape(2,4)#[0,1,2,3],[4,5,6,7]

p=re_sp[1, 2]
#[row(index from 0),column(index from 0)]
#[0,1,2,3]
#[7,5,6,4](here come 1 no row)
# 0 1 2 3---index position (then come 2 no column index from 0)= 6 value
print("reshape:\n",re_sp)
print("\n")
print("m[1, 2] is:\n",p)
print(" \n")
#m[1,2] is 6

#23.Given the same m, what is m[:, :3]?
m = np.arange(8).reshape(2,4)
print("m[:, :3] is:\n",m[:, :3])
print("\n")
#[[0 1 2]
 # [4 5 6]]

#24.Given data = np.arange(8), what does data > 5 produce?
data = np.arange(8)
print(data)
mask=data>5
print(mask)
print("\n")

#25.What is the result of data[data > 5] if data = np.arange(8)
data = np.arange(8)
a= data[data > 5]
print("The value of data[data > 5]\n:",a)#means all value greater than 5
print("\n")

#26 Given a 4×4 matrix called matrix, use np.ix_ to select rows 1, 2, 3 and columns 0, 2, 3.
a= np.arange(16).reshape(4,4)
result =a[np.ix_([1,2,3], [0,2,3])]
print("Output is:\n",result)
```
